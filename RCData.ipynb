```python
cashCounter = 0
americanCounter = 0
bankCounter = 0
masterCounter = 0 
visaCounter = 0
for i in range(0, len(userPayment)):
    if userPayment.iloc[i]['Upayment'] == 'cash':
        cashCounter = cashCounter +1
        #print (userPayment.iloc[i]['userID'])
    elif userPayment.iloc[i]['Upayment'] == 'American_Express':
        americanCounter = americanCounter +1
        #print (userPayment.iloc[i]['userID'])
    elif userPayment.iloc[i]['Upayment'] == 'bank_debit_cards':
        bankCounter = bankCounter +1
        #print (userPayment.iloc[i]['userID'])
    elif userPayment.iloc[i]['Upayment'] == 'MasterCard-Eurocard':
        masterCounter = masterCounter +1
        #print (userPayment.iloc[i]['userID'])
    elif userPayment.iloc[i]['Upayment'] == 'VISA':
        visaCounter = visaCounter +1
        #print (userPayment.iloc[i]['userID'])

#print percentage of people who use cash
print(100*(cashCounter / len(userPayment)))
#print percentage of people who use American Express
print(100*(americanCounter / len(userPayment)) ) 
#print percentage of people who  bank or debit cards
print(100*(bankCounter / len(userPayment)) ) 
#print percentage of people who use master cards
print(100*(masterCounter / len(userPayment)) ) 
#print percentage of people who use visa cards
print(100*(visaCounter / len(userPayment)) ) 
```

```python
maxCuisine = ''

counter = 0
maximumCount = 0

for i in range(0, len(userCuisine)):
    for y in range(0, len(userCuisine)):
        if userCuisine.iloc[i]['Rcuisine'] == userCuisine.iloc[y]['Rcuisine']:
            counter = counter + 1
    
    if counter > maximumCount:
        maxCuisine = userCuisine.iloc[i]['Rcuisine']
        maximumCount = counter
    counter = 0
        
        
print(maxCuisine + " with count: ")
maximumCount
```
